```r
library(repr)
data(faithful)
head(faithful)
```

```r
faithful$eruptions_rescale = scale(faithful$eruptions)
faithful$waiting_rescale = scale(faithful$waiting)
summary(faithful)

edist <- dist(faithful[,3:4])
#edist
```

```r
sclust <- hclust(edist, method="single")
cclust <- hclust(edist, method="complete")

par(mfrow=c(1:2))
options(repr.pot.width=8, repr.plot.height=4)
plot(sclust)
plot(cclust)
```

```r
comp_clust <- cutree(cclust, k=2)
table(comp_clust)

plot(faithful$eruptions, faithful$waiting, col=comp_clust)
```
