## Pittsburgh Police Arrest Data

```r
library(repr)
```

```r
arrest <- read.csv("arrest.csv")
head(arrest)
```

### Variable Description: 
- **PK**: ARREST_ID
- **CCR**: Incident Number
- **AGE**: Age of suspect
- **GENDER**: Gender of suspect
- **RACE**: Race of suspect
- **ARRESTTIME**: Date and Time of arrest
- **ARRESTLOCATION**: Location of Arrest
- **OFFENSES**: All offenses listed for the incident
- **INCIDENTLOCATION**: Incident Location
- **INCIDENTNEIGHBORHOOD**: Neighborhood where incident occured
- **INCIDENTZONE**: Police Zone where Incident occurred
- **INCIDENTTRACT**: 
- **COUNCIL_DISTRICT**: Council district, from 1-9
- **PUBLIC_WORKS_DIVISION**: Work of building such things as roads, schools, and reservoirs, carried out by the government for the community, from 1-6
- **X**: The geocoded X coordinate of the incident location (using county geocoder)
- **Y**: The geocoded Y coordinate of the incident location (using county geocoder)
#### Removing missing values (NA):

```r
arrest <- arrest[complete.cases(arrest),]
summary(arrest)
```

### Demonstracte the distribution of suspects by Race, Gender and Age Range

```r
par(mfrow=c(2:1))
options(repr.plot.width=4, repr.plot.height=8)
plot(arrest$GENDER, xlab="Gender", ylab="Distribution of Suspects", main="Suspects by Gender")
plot(arrest$RACE, xlab="Race", ylab="Distribution of Syuspects", main="Suspects by Race")
```

```r
# # of male suspects
table(arrest$GENDER=="M")
# # of Black/ white suspects
table(arrest$RACE=="B")
table(arrest$RACE=="W")
```
