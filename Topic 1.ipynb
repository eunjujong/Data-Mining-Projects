## Birthweight Data

```r
lbw <- read.csv("Infantslbw.csv")
head(lbw)
```

### Histogram: Birthweight vs. Distribution of women

```r
hist(lbw$bwt, xlab="Birth Weight", ylab="Distr. of Women", xlim=c(500,5000), ylim=c(0,50))
```

### Histogram: Age of Pregnancy vs. Distribution of women

```r
hist(lbw$age, xlab="Age", ylab="Distr. of Women", xlim=c(10,45), ylim=c(0,70))
```

### Scatterplot: Age of Pregnancy vs. Berthweight

```r
plot(lbw$age, lbw$bwt, xlab="Age", ylab="Birth Weights", xlim=c(10,45), ylim=c(700,5000))
```

### Scatterplot: weight at last menstrual period: 80-250 lb vs. Birthweight

```r
plot(lbw$lwt, lbw$bwt, xlab="Weights at Last Menstrual Period", ylab="Birth Weights", xlim=c(80,250), ylim=c(700,5000))
```

```r
lbw$race <- factor(c("White", "Black", "Other"))
levels(lbw$race) <- list("White"="white", "Black"="Black", "Other"="Other")
```

### Boxplot: Race vs. Birthweight

```r
plot(lbw$race, lbw$bwt, ylab="Beirth Weights")
```

### Boxplot: false premature labor vs. Birthweight

```r
lbw$ptl <- as.factor(lbw$ptl)
plot(lbw$ptl, lbw$bwt, xlab="False Premature Labor", ylab="Birthweight")
```

### A new variable indicating whether there have been any instances of false premature labor: biptl

```r
lbw$biptl <- lbw$ptl > 0
head(lbw)
```

```r
lbw$biptl <- as.factor(lbw$biptl)
plot(lbw$biptl, lbw$bwt, xlab="Instances of False Premature Labor", ylab="Birthweight")
```
